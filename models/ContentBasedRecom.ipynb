```python
import pandas as pd
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')

path="/content/drive/MyDrive/Colab Notebooks/Online_courses_final1_updated.csv"
course =pd.read_csv(path)
```

```python
course.head()
```

```python
course['tag'] = course['Category'] +' '+ course['Sub-Category'] +' '+ course['Skills']
```

```python
course.head()
```

```python
new_df = course[['CourseID', 'Title', 'tag']]
```

```python
new_df.head()
```

```python
# Assuming 'new_df' is your DataFrame
new_df = new_df.drop_duplicates(subset=['Title'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vec = cv.fit_transform(new_df['tag'].values.astype('U')).toarray()
```

```python
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim = cosine_similarity(vec)
```

```python
sim
```

```python
new_df[new_df['Title']== 'Machine Learning Specialization']
```

```python
dist = sorted(list(enumerate(sim[0])), reverse=True, key=lambda vec: vec[1])
```

```python
dist
```

```python
for i in dist[0:10]:
  print(new_df.iloc[i[0]].Title)
```

```python
def recommend(course):
   index = new_df[new_df['Title']== course].index[0]
   dist = sorted(list(enumerate(sim[index])), reverse=True, key=lambda vec: vec[1])
   for i in dist[0:10]:
     print(new_df.iloc[i[0]].Title)
```

```python
recommend('Introduction to Data Science Specialization')
```
