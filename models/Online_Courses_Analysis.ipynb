```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://drive.google.com/uc?export=download&id=1yxv5F0pDqat6kpYeqxy_kdtWl4lLcQ8N'
data = pd.read_csv(url)
data.head()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Number of viewers', y='Rating', hue='Category', alpha=0.7)

plt.xscale('log')  # Apply logarithmic scale to x-axis
plt.title('Ratings vs. Number of Viewers')
plt.xlabel('Number of Viewers (Log Scale)')
plt.ylabel('Rating')
plt.xticks(rotation=45)

plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Bar Graph: Average Rating per Course Type
plt.figure(figsize=(12, 6))
avg_ratings = data.groupby('Course Type')['Rating'].mean().sort_values()
avg_ratings.plot(kind='barh', color='salmon')
plt.title('Average Rating per Course Type')
plt.xlabel('Average Rating')
plt.ylabel('Course Type')
plt.show()
```

```python
# Pie Chart: Distribution of Categories
plt.figure(figsize=(8, 8))
category_counts = data['Category'].value_counts()
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Courses by Category')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()
```

```python
# Box Plot: Course Ratings by Category
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Category', y='Rating', palette='Set2')
plt.title('Course Ratings by Category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count Plot: Number of Courses by Language
plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Language', palette='viridis', order=data['Language'].value_counts().index)
plt.title('Number of Courses by Language')
plt.xlabel('Language')
plt.ylabel('Number of Courses')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = data.corr(numeric_only=True)  # Added numeric_only=True to handle non-numeric columns
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.show()
```
